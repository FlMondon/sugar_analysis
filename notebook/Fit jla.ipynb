# Plot JLA Light Curves


```python
sn_good = []
sn_bad = [] 
sn_fail = []
dic = pkl.load(open('../../sugar_analysis_data/resfitlc_SDSS_sugarq1q3.pkl'))

cut = 5.
for sn_name in dic['data'].keys():
    try: 
        if float(dic['data'][sn_name]['res']['parameters'][3]) >= -cut and float(dic['data'][sn_name]['res']['parameters'][3])<= cut:
            sn_good.append(sn_name)
        else :
            sn_bad.append(sn_name)
    except:
        sn_fail.append(sn_name)

print len(dic['data']), len(sn_good),  len(sn_bad),  len(sn_fail)
```
