# Plot results fit spectro/photo

```python
errors = np.zeros_like(rids.params)
spec_par = np.zeros_like(rids.params)
param_name = ['grey','q1', 'q2', 'q3', 'Av']
for i, sn_name in enumerate(rids.dic_res):
    for j in range(5):
        if j==0:
            spec_par[i,j] = spec_param[sn_name][param_name[j]]+sa.distance_modulus_th(rids.dic_res[sn_name]['zcmb'],rids.dic_res[sn_name]['zhl'])
            errors[i,j] = np.sqrt(rids.dic_res[sn_name]['res']['covariance'][j+1,j+1])/(2.5 * np.log(10) * (10**(-0.4*rids.params[i,0])))
        else:
            spec_par[i,j] = spec_param[sn_name][param_name[j]] 
            errors[i,j] = rids.dic_res[sn_name]['res']['covariance'][j+1,j+1]
print errors
```

```python
plt.figure(figsize=(20, 20))
rcParams['font.size'] = 20.
font = {'family': 'normal', 'size': 20}
rc('axes', linewidth=3)
rc("text", usetex=True)
rc('font', family='serif')
rc('font', serif='Times')
rc('legend', fontsize=25)
rc('xtick.major', size=5, width=3)
rc('ytick.major', size=5, width=3)
rc('xtick.minor', size=3, width=3)
rc('ytick.minor', size=3, width=3)
LIM = [0, 8,8,6]
result_path = '../../sugar_analysis_data/results/save_plot/'
plt.figure(figsize=(15, 25))
for i in range(5):
    plt.subplot(5, 2, 2*i+1)
    if 2*i+1 < 10:
        plt.plot([-100.,100.],[-100.,100.],color='r')
        plt.ylabel(param_name[i])
        plt.errorbar(spec_par[:,i], rids.params[:, i], yerr=errors[:, i], fmt='.')
        cor = Statistics.correlation_weighted(spec_par[:,i], rids.params[:, i],w=1./errors[:, i])
        if 2*i+1 == 1:
            plt.text(36.5, 37.6, round(cor,3) , fontweight = 'bold', fontsize = 20)
            plt.ylim(35.5,38.5)
            plt.xlim(35.5,38.5) 
        if 2*i+1 <= 7 and 2*i+1>=2:
            plt.text(-LIM[i]+1.5, LIM[i]-1.5, round(cor,3) , fontweight = 'bold', fontsize = 20)
            plt.ylim(-LIM[i],LIM[i])
            plt.xlim(-LIM[i],LIM[i])
        if 2*i+1 == 9:
            plt.text(-0.5, 0.6, round(cor,3) , fontweight = 'bold', fontsize = 20)
            plt.ylim(-1,1.)
            plt.xlim(-1.,1.)   
        plt.subplot(5, 2, 2*(i+1))
        plt.hist(spec_par[:,i]-rids.params[:, i],20, label='$\Delta$ '+param_name[i]) 
        plt.legend()
        plt.xlim(-5,5)

pngfile = result_path+'results_specvsphot.png'
plt.savefig(pngfile, bbox_inches='tight')        
plt.show()
```
