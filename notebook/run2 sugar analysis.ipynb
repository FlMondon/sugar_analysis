## Plot results fit spectro/photo

```python
errors = np.zeros_like(rids.params)
spec_par = np.zeros_like(rids.params)
param_name = ['grey','q1', 'q2', 'q3', 'Av']
for i, sn_name in enumerate(rids.dic_res):
    for j in range(5):
        if j==0:
            spec_par[i,j] = spec_param[sn_name][param_name[j]]+sa.distance_modulus_th(rids.dic_res[sn_name]['zcmb'],rids.dic_res[sn_name]['zhl'])
            errors[i,j] = np.sqrt(rids.dic_res[sn_name]['res']['covariance'][j+1,j+1])/(2.5 * np.log(10) * (10**(-0.4*rids.params[i,0])))
        else:
            spec_par[i,j] = spec_param[sn_name][param_name[j]] 
            errors[i,j] = rids.dic_res[sn_name]['res']['covariance'][j+1,j+1]
plt.figure(figsize=(20, 20))
rcParams['font.size'] = 20.
font = {'family': 'normal', 'size': 20}
rc('axes', linewidth=3)
rc("text", usetex=True)
rc('font', family='serif')
rc('font', serif='Times')
rc('legend', fontsize=25)
rc('xtick.major', size=5, width=3)
rc('ytick.major', size=5, width=3)
rc('xtick.minor', size=3, width=3)
rc('ytick.minor', size=3, width=3)
LIM = [0, 8,8,6]
result_path = '../../sugar_analysis_data/results/save_plot/'
plt.figure(figsize=(15, 25))
for i in range(5):
    plt.subplot(5, 2, 2*i+1)
    if 2*i+1 < 10:
        plt.plot([-100.,100.],[-100.,100.],color='r')
        plt.ylabel(param_name[i])
        plt.errorbar(spec_par[:,i], rids.params[:, i], yerr=errors[:, i], fmt='.')
        cor = Statistics.correlation_weighted(spec_par[:,i], rids.params[:, i],w=1./errors[:, i])
        rms = np.std(spec_par[:,i]-rids.params[:, i])
        rms_err = np.std(spec_par[:,i]-rids.params[:, i])/(2*np.sqrt(len(rids.params[:, i])))
        mean = np.mean(spec_par[:,i]-rids.params[:, i])
        mean_err = np.std(spec_par[:,i]-rids.params[:, i])/np.sqrt(len(rids.params[:, i]))
        if 2*i+1 == 1:
            plt.text(35.6, 37.9, 'mean = '+str(round(mean,3))+' +-'+ str(round(mean_err,3)), fontweight = 'bold', fontsize = 25)
            plt.text(35.6, 37.3, 'rms = '+str(round(rms,3))+' +-'+ str(round(rms_err,3)), fontweight = 'bold', fontsize = 25)
            plt.ylim(35.5,38.5)
            plt.xlim(35.5,38.5) 
        if 2*i+1 <= 7 and 2*i+1>=2:
            plt.text(-LIM[i]+1.5, LIM[i]-2.5, 'mean = '+str(round(mean,3))+' +-'+ str(round(mean_err,3)), fontweight = 'bold', fontsize = 25)
            plt.text(-LIM[i]+1.5, LIM[i]-4, 'rms = '+str(round(rms,3))+' +-'+ str(round(rms_err,3)), fontweight = 'bold', fontsize = 25)
            plt.ylim(-LIM[i],LIM[i])
            plt.xlim(-LIM[i],LIM[i])
        if 2*i+1 == 9:
            plt.text(-0.8, 0.45, 'mean = '+str(round(mean,3))+' +-'+ str(round(mean_err,3)), fontweight = 'bold', fontsize = 25)
            plt.text(-0.8, 0.3, 'rms = '+str(round(rms,3))+' +-'+ str(round(rms_err,3)), fontweight = 'bold', fontsize = 25)
            plt.ylim(-1,1.)
            plt.xlim(-1.,1.)   
        plt.subplot(5, 2, 2*(i+1))
        plt.hist(spec_par[:,i]-rids.params[:, i],20, label='$\Delta$ '+param_name[i]) 
        plt.legend()
        #plt.xlim(-5,5)

pngfile = result_path+'results_specvsphotBVR.png'
plt.savefig(pngfile, bbox_inches='tight')        
plt.show()
```

# LC residuals

```python
nb_bin_phase = 9
dic =  pkl.load(open('../../sugar_analysis_data/resfitlc_SNf_5new_fU_10new_fI_10_sugar.pkl'))
dic_res  = dic['data']
fit_exclude = []
dic_del = {}
spec_param_path = '../../sugar_model/sugar_parameters.pkl'
spec_param = pkl.load(open(spec_param_path))

for sn_name in dic_res.keys():
    if dic_res[sn_name]['res'] != 'fit fail' and sn_name in spec_param.keys():
        dic_del[sn_name] = dic_res[sn_name]
    else:
        fit_exclude.append(sn_name)
dic_res = dic_del 
```

```python
source = sncosmo.get_source('sugar')
dust = sncosmo.CCM89Dust()
model = sncosmo.Model(source=source, 
                  effects=[dust], 
                  effect_names=['mw'], 
                  effect_frames=['obs']) 
snf_sys = sncosmo.get_magsystem('vega_snf_10')
residue = {}
results_sugar = { 'new_fI_10' : {'residue':[], 'data_err':[],'res_bin_phase':{}},
                   'fR_10' : {'residue':[], 'data_err':[],'res_bin_phase':{}},
                   'fV_10' : {'residue':[], 'data_err':[],'res_bin_phase':{}},
                   'fB_10' : {'residue':[], 'data_err':[],'res_bin_phase':{}},
                   'new_fU_10' : {'residue':[], 'data_err':[],'res_bin_phase':{}}}

phase = np.linspace(-12,52,nb_bin_phase)
for band in results_sugar.keys():
    for i in range(len(phase)-1):
        results_sugar[band][str(phase[i])+'_'+str(phase[i+1])] = []
        
        
for sn_name in dic_res.keys():
    tmax = float(dic_res[sn_name]['res']['parameters'][1])
    model.set(z=float(dic_res[sn_name]['res']['parameters'][0]))
    model.set(t0 = tmax)
    model.set(Xgr = float(dic_res[sn_name]['res']['parameters'][2]))
    model.set(q1 = float(dic_res[sn_name]['res']['parameters'][3]))
    model.set(q2 = float(dic_res[sn_name]['res']['parameters'][4]))
    model.set(q3 = float(dic_res[sn_name]['res']['parameters'][5]))
    model.set(A = float(dic_res[sn_name]['res']['parameters'][6]))
    model.set(mwebv = float(dic_res[sn_name]['res']['parameters'][7]))
    for i, band in enumerate(dic_res[sn_name]['data_table']['band']):
        time = dic_res[sn_name]['data_table']['time'][i]
        try:
            model_mag = model.bandmag(band,'vega_snf_10', time)
        except:
            continue
        if time >= tmax - 12. and time <= tmax + 48.:
            try:
                mag = snf_sys.band_flux_to_mag(dic_res[sn_name]['data_table']['flux'][i], band)
            except:
                continue
            results_sugar[band]['residue'].append(mag - model_mag)
            err_mag = dic_res[sn_name]['data_table']['fluxerr'][i] /dic_res[sn_name]['data_table']['flux'][i] *1.0857362047581294
            results_sugar[band]['data_err'].append(err_mag)

            for i in range(len(phase)-1):
                if time-tmax >= phase[i] and time-tmax <=phase[i+1]:
                    results_sugar[band][str(phase[i])+'_'+str(phase[i+1])].append(mag - model_mag)
                    continue
```
